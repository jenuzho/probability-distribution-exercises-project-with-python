# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

# **DISTRIBUCIÓN NORMAL**

```python
from scipy.stats import norm

# Define all the probability functions related to the normal distribution

def dnorm(x, mean=0, std=1):
    """
    Calculates the density (PDF) of the normal distribution.
    
    Parameters:
        x (float or array-like): Point(s) at which to evaluate the density.
        mean (float): Mean (μ) of the normal distribution. Default is 0.
        std (float): Standard deviation (σ) of the normal distribution. Default is 1.
    
    Returns:
        float or array-like: Density value(s) at the given point(s).
    """
    if std <= 0:
        raise ValueError("Standard deviation (std) must be greater than 0.")
    
    result = norm.pdf(x, loc=mean, scale=std)
    return result


def pnorm(x, mean=0, std=1):
    """
    Calculates the cumulative distribution function (CDF) of the normal distribution.
    
    Parameters:
        x (float or array-like): Point(s) at which to evaluate the CDF.
        mean (float): Mean (μ) of the normal distribution. Default is 0.
        std (float): Standard deviation (σ) of the normal distribution. Default is 1.
    
    Returns:
        float or array-like: Cumulative probability up to the given point(s).
    """
    if std <= 0:
        raise ValueError("Standard deviation (std) must be greater than 0.")
    
    result = norm.cdf(x, loc=mean, scale=std)
    return result


def qnorm(p, mean=0, std=1):
    """
    Calculates the quantile (inverse CDF) for the normal distribution.
    
    Parameters:
        p (float or array-like): Probability value(s) (0 <= p <= 1).
        mean (float): Mean (μ) of the normal distribution. Default is 0.
        std (float): Standard deviation (σ) of the normal distribution. Default is 1.
    
    Returns:
        float or array-like: Quantile corresponding to the given probability.
    """
    if not (0 <= p <= 1):
        raise ValueError("Probability (p) must be between 0 and 1.")
    if std <= 0:
        raise ValueError("Standard deviation (std) must be greater than 0.")
    
    result = norm.ppf(q=p, loc=mean, scale=std)
    return result


def rnorm(n, mean=0, std=1):
    """
    Generates random variables from the normal distribution.
    
    Parameters:
        n (int): Number of random variables to generate.
        mean (float): Mean (μ) of the normal distribution. Default is 0.
        std (float): Standard deviation (σ) of the normal distribution. Default is 1.
    
    Returns:
        array-like: Array of random variables from the normal distribution.
    """
    if n <= 0 or not isinstance(n, int):
        raise ValueError("Number of samples (n) must be a positive integer.")
    if std <= 0:
        raise ValueError("Standard deviation (std) must be greater than 0.")
    
    result = norm.rvs(size=n, loc=mean, scale=std)
    return result
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define probability functions
def dnorm(x, mean=0, std=1):
    """
    Calculates the density (PDF) of the normal distribution.
    """
    if std <= 0:
        raise ValueError("Standard deviation (std) must be greater than 0.")
    return norm.pdf(x, loc=mean, scale=std)

def pnorm(x, mean=0, std=1):
    """
    Calculates the cumulative distribution function (CDF) of the normal distribution.
    """
    if std <= 0:
        raise ValueError("Standard deviation (std) must be greater than 0.")
    return norm.cdf(x, loc=mean, scale=std)

def qnorm(p, mean=0, std=1):
    """
    Calculates the quantile (inverse CDF) for the normal distribution.
    """
    p = np.asarray(p)  # Ensure p is array-like
    if np.any((p < 0) | (p > 1)):
        raise ValueError("All probabilities (p) must be between 0 and 1.")
    if std <= 0:
        raise ValueError("Standard deviation (std) must be greater than 0.")
    return norm.ppf(q=p, loc=mean, scale=std)

def rnorm(n, mean=0, std=1):
    """
    Generates random variables from the normal distribution.
    """
    if n <= 0 or not isinstance(n, int):
        raise ValueError("Number of samples (n) must be a positive integer.")
    if std <= 0:
        raise ValueError("Standard deviation (std) must be greater than 0.")
    return norm.rvs(size=n, loc=mean, scale=std)

# Parameters
mean = 0
std = 1

# Generate data for visualization
x = np.linspace(-5, 5, 1000)
pdf = dnorm(x, mean, std)  # PDF values
cdf = pnorm(x, mean, std)  # CDF values
probabilities = np.linspace(0.001, 0.999, 1000)
quantiles = qnorm(probabilities, mean, std)  # Quantile values

# Plotting
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# PDF
axs[0].plot(x, pdf, 'r-', label='PDF')
axs[0].set_title('Probability Density Function (PDF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('Density')
axs[0].legend()

# CDF
axs[1].plot(x, cdf, 'b-', label='CDF')
axs[1].set_title('Cumulative Distribution Function (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('Cumulative Probability')
axs[1].legend()

# Quantiles
axs[2].plot(probabilities, quantiles, 'g-', label='Quantile Function (PPF)')
axs[2].set_title('Quantile Function (PPF)')
axs[2].set_xlabel('Probability')
axs[2].set_ylabel('Quantile')
axs[2].legend()

plt.tight_layout()
plt.show()

```

### DISTRIBUCIÓN UNIFORME

```python
from scipy.stats import uniform

# Uniform
def dunif(x, low = 0, high = 1):
    """
    Calculates the point estimate of the uniform distribution
    """
    result = uniform.pdf(x, loc = low, scale = (high - low))

    return result

def punif(q, low = 0, high = 1):
    """
    Calculates the cumulative of the uniform distribution
    """
    result = uniform.cdf(q, loc = low, scale = (high - low))

    return result

def qunif(p, low = 0, high = 1):
    """
    Calculates the quantile function from the uniform distribution
    """
    result = uniform.ppf(p, loc = low, scale = (high - low))

    return result

def runif(n, low = 0, high = 1):
    """
    Generates random variables from the uniform distribution
    """
    result = uniform.rvs(loc = low, scale = (high - low), size = n)

    return result
```

```python
np.random.seed(42)

low = 0
high = 1
data = np.arange(-5, 5, 0.01)

pmf = dunif(data, low = low, high = high)
cdf = punif(data, low = low, high = high)
ppf = qunif(data, low = low, high = high)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].plot(data, pmf, "y-")
axis[1].plot(data, cdf, "r-")
axis[2].plot(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
